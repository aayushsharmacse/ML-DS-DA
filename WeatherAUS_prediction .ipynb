Before we process anything we see that more than half of the data corresponding to the "Evaporation","Sunshine","Cloud9am","Cloud3pm" features are missing and we already have and will get more features by breaking , encoding other features later on. So it is better to drop whole of these features now early on.

```python
df=df.drop(["Evaporation","Sunshine","Cloud9am","Cloud3pm"],axis=1)
df
```

Here are some missing values in other features as well , we can simply remove the corresponding rows because we have a large amount of data present.

```python
df.dropna(inplace=True)
df
```

```python
df.duplicated().value_counts()
```

This means our data was safe from repetition.

Analysing Date Feature

```python
df["Date"].head()
```

As we can see dates is present as object so we need to convert it into datatime object so we can futher process it and break it into more useful features .

```python
df["Date"] = pd.to_datetime(df["Date"],
 format = "%Y-%m-%d",
 errors = "coerce")
```

```python
df.Date.head()
```

Breaking the datatime object into some useful features such as day , month , year

```python
df["Day"]=df.Date.dt.day
df["Month"]=df.Date.dt.month
df["Year"]=df.Date.dt.year
```

We can now see our new features in the end of dataframe.

```python
df.head()
```

Now we don't need "Date" since we already have its replacements so we can drop it.

```python
df=df.drop(["Date"],axis=1)
df.head()
```

Lets check some information about updated data.

```python
df.describe()
```

We see our data especially "Rainfall" , "Evaporation","WindSpeed9am" are having outliners definitely as can be seen from the difference between the 75th percentile data comparing to the max values. Even the other way round "Humidity9am" seems to have significant difference between the 25th percentile and min value.

So , we should remove rows corresponding to the outliners.

```python
dfx=df.describe()
def remove_outliners():
  for i in dfx.columns:
    # if(i=="Day" or i=="Month" or i=="Year"):
        # continue;

    print(i,dfx[i]["mean"])
    print("z for",i,3*dfx[i]["std"]+dfx[i]["mean"])
    df.drop((df[df[i]>3*dfx[i]["std"]+dfx[i]["mean"]]).index,inplace=True,axis=0)
    # mean=dfx[i]["mean"];
    # std=dfx[i]["std"]


remove_outliners()
```

Encoding the data to make the categorical features as numerical so that we can train it in our model.

```python
encoded_data=pd.get_dummies(df,columns=['WindDir9am','WindDir3pm',"Location","WindGustDir","RainToday","RainTomorrow"],drop_first=True)
```

```python
encoded_data.head()
```

Separating the input (x) and output (y).

```python
x=encoded_data.drop(["RainTomorrow_Yes"],axis=1)
```

```python
x.head()
```

```python
y=encoded_data["RainTomorrow_Yes"]
```

```python
y.head()
```

```python
y.value_counts()
```

Spliting the data into training and test data.

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

Since the differences between the values are big i shall standardize the data as well.

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
xtrain=st.fit_transform(xtrain)
```

```python
xtest=st.transform(xtest)
```

Accuracy Measures

```python
from sklearn.metrics import accuracy_score
```

Model-Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
```

```python
y_dtc_predict=dtc.predict(xtest)
```

```python
accuracy_score(y_dtc_predict,ytest)
```

SVM
