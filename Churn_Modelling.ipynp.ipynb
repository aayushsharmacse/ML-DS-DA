```python
encoded_data=pd.get_dummies(df,columns=['Gender','Geography'],drop_first=True)
```

```python
#drop_first shows 1 less column
#eg Female & Male show only Male here , female already know if not male
#eg2 here 3 countrys france,germany,spain but it will drop 1st i.e france if we know two country 0 0 then third is 
#already know 1 see eg below
#female   male
#  0        1
#  1        0
#if we only need to know female we can already know male by that alone
#female                     Male??if not female
#  0
#  1
#simlarly for countries
# france       germany         spain
#   0             0              1
#   1             0              0
#   0             1              0
# if only know 2 countrys we can tell 3rd ourselves
```

```python
x=encoded_data.drop(['Exited'],axis=1)
```

```python
y=encoded_data['Exited']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
xtrain=st.fit_transform(xtrain)
```

```python
xtest=st.transform(xtest)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
```

```python
gb.fit(xtrain,ytrain)
```

```python
y_gb_predict=gb.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_gb_predict,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
y_lr_pred=lr.predict(xtest)
```

```python
accuracy_score(y_lr_pred,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc=KNeighborsClassifier()
```

```python
knc.fit(xtrain,ytrain)
```

```python
y_knc_pred=knc.predict(xtest)
```

```python
accuracy_score(y_knc_pred,ytest)
```
